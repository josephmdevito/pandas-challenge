```python
# dependencies and setup of environment
import pandas as pd
from pathlib import Path
import os

# load csv data
school_data_csv = Path(r"C:\Users\joemd\Desktop\pandas-challenge\Starter_Code\PyCitySchools\Resources\schools_complete.csv")
student_data_csv = Path(r"C:\Users\joemd\Desktop\pandas-challenge\Starter_Code\PyCitySchools\Resources\students_complete.csv")

# read data files and store into Pandas DataFrames
school_data = pd.read_csv(school_data_csv)
student_data = pd.read_csv(student_data_csv)

# combine the data into a single datset
data_merge = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
data_merge.head()
```

```python
# start DISTRICT SUMMARY
# calculate the total number of unique schools in the district
school_count = school_data['school_name'].count()
print(school_count)
```

```python
# calculate the total number of students in the district
students_count = student_data['student_name'].count()
print(students_count)
```

```python
# calculate the total budget for the district
total_budget = school_data['budget'].sum()
print(total_budget)
```

```python
# calculate the average math score in the district
math_avg = data_merge['math_score'].mean()
# round to second decimal and print result
math_avg_round = round(math_avg, 2)
print(math_avg_round)
```

```python
# calculate the average reading score in the district
read_avg = data_merge['reading_score'].mean()
# round to second decimal and print result
read_avg_round = round(read_avg, 2)
print(read_avg_round)
```

```python
# calculate the % of student who passed reading in the district
read_pass = data_merge.loc[data_merge['reading_score'] >= 70]['reading_score'].count()
print(read_pass)
read_pass_per = (read_pass/students_count) * 100
# round to second decimal and print result 
read_pass_per_round = round(read_pass_per, 2)
print(read_pass_per_round)
```
